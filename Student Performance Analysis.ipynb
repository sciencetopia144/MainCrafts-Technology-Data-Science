### Step 3: Core Data Analysis
In this section, we address the primary research questions defined in the project guidelines to extract meaningful insights from the student dataset:

* **Academic Benchmarking:** We calculate the average final grade (G3) to establish a baseline for overall student performance.
* **High-Achiever Identification:** We filter the data to determine the number of students who demonstrated exceptional performance by scoring above 15.
* **Study Habit Impact:** We perform a correlation analysis to investigate if increased study time directly translates to better academic results.
* **Demographic Performance Trends:** We compare performance metrics by gender to identify if any significant disparities exist in average final scores.

```python
# 1. Average final grade (G3)
avg_g3 = df['G3'].mean()
print("Average Final Grade (G3):")
print(avg_g3)
```

```python
# 2. How many students scored above 15?
# (Assuming 'G3' is the grade column)
students_above_15 = df[df['G3'] > 15].shape[0]
print("Number of students scoring above 15:")
print(students_above_15)
```

```python
# 3. Is study time correlated with performance?
# We calculate the correlation coefficient between 'studytime' and 'G3'
correlation = df['studytime'].corr(df['G3'])
print(f"Correlation between study time and performance: {correlation:.2f}")
print("(" , correlation, ")" ,  "to be exact")
```

```python
# 4. Which gender performs better on average?
gender_performance = df.groupby('sex')['G3'].mean()
print("Average G3 score by gender:")
print(gender_performance)
```

```python
# Identifying the better-performing gender
better_gender = gender_performance.idxmax()
print(f"\nOn average, the {better_gender} gender performs better.")
```

### üìä Analysis Methodology

* **Average Final Grade:** We utilize the `.mean()` function on the **G3** column to determine the overall academic baseline for the student cohort.
* **High Performance Count:** By applying a **Boolean filter** `df['G3'] > 15`, we isolate the high-achieving students and use `.shape[0]` to quantify this specific group.
* **Correlation Analysis:** The `.corr()` method is used to measure the statistical relationship between study habits and grades; a positive coefficient suggests that increased study time is associated with higher performance.
* **Comparative Analysis:** We use the `.groupby()` function to aggregate data by **Gender**, allowing for a direct comparison of average academic outcomes between different demographic groups.
* **Peak Value Identification:** The `.idxmax()` function is used to automatically extract the specific label (e.g., 'M' or 'F') associated with the highest average score calculated during the grouping process.

### Step 4: üìà Exploratory Data Analysis: Visualizations
To better understand the relationships within our dataset, we have generated three key visualizations for

* **Grade Distribution:** A **Histogram** with a Kernel Density Estimate (KDE) is used to visualize the spread of final grades (G3) and identify common performance clusters.
* **Impact of Study Time:** A **Scatterplot** maps study hours against final grades to visually assess if there is a linear correlation between academic effort and results.
* **Gender-Based Comparison:** A **Bar Chart** displays the average scores for male and female students, allowing for a quick comparison of performance trends across demographic groups.

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

```python
#We set the visual style as white-grid
sns.set_theme(style="whitegrid")
```

```python
# lets create a figure with subplots to organize your charts
plt.figure(figsize=(15, 10))
```

**1. Histogram of grades (G3)**


```python
plt.subplot(2, 2, 1)
sns.histplot(df['G3'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Grade')
plt.ylabel('Frequency')
```

**2. Scatterplot: Study Time vs Grades**

```python
plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='studytime', y='G3', alpha=0.6, color='coral')
plt.title('Study Time vs. Final Grades')
plt.xlabel('Weekly Study Time (Hours)')
plt.ylabel('Final Grade (G3)')
```

**3. Bar Chart: Male vs Female Average Score**

```python
# 3. Bar Chart: Male vs Female Average Score 
plt.subplot(2, 2, 3)
sns.barplot(data=df, x='sex', y='G3', estimator='mean', hue='sex', palette='pastel')
plt.title('Average Final Score by Gender')
plt.xlabel('Gender (F: Female, M: Male)')
plt.ylabel('Average Grade')
```

**4. Adjust layout to prevent overlap**

```python
plt.tight_layout()
plt.show()
```
